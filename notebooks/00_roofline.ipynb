# NB-0 · Roofline: are parallel tokens free?

**Claim under test:** single-stream decode is memory-bandwidth-bound, so a forward pass over W new tokens costs ~the same as W=1 up to a surprisingly large W. If the cost curve stays flat to W≈32–64, block-parallel generation gets its tokens "for free" and the whole project's arithmetic stands.

**How to read the output:** left panel = ms per pass vs W (flat = good). Right panel = cost ratio vs W=1, with the compute-bound diagonal for contrast. Runtime: ~3–5 min on a T4.

Runtime → Change runtime type → **GPU** before running.

```python
# If the repo is private, use: pip install git+https://<TOKEN>@github.com/mccooking/blockspeed.git
%pip -q install git+https://github.com/mccooking/blockspeed.git

import torch
assert torch.cuda.is_available(), "No GPU: Runtime -> Change runtime type -> GPU"
print(torch.cuda.get_device_name(0))
```

```python
from blockspeed.roofline import measure, verdict

result = measure("Qwen/Qwen3-0.6B")  # ~3-5 min on a T4
print()
print(verdict(result))
```

```python
from blockspeed.roofline import plot, save_result

path = save_result(result, results_dir="results")
print("saved:", path)
plot(result)
```
